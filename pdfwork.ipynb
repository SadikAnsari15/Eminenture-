```python
import camelot
import pandas as pd

# Extract tables from all pages of the PDF using 'stream' flavor
tables = camelot.read_pdf('natwest-1.pdf', pages='all', flavor='stream')

# Check the number of tables extracted
print(f"Total tables extracted: {len(tables)}")

# Initialize an empty list to store DataFrames from each table
dfs = []

# Loop through all the tables and convert them into DataFrames
for table in tables:
    df = table.df  # Convert each table into a DataFrame
    df_cleaned = df.dropna(how='all', axis=0)  # Drop rows where all values are NaN
    df_cleaned = df_cleaned.dropna(how='all', axis=1)  # Drop columns where all values are NaN
    df_cleaned.reset_index(drop=True, inplace=True)  # Reset the index
    dfs.append(df_cleaned)  # Append the cleaned DataFrame to the list

# Combine all the DataFrames into a single DataFrame
if dfs:
    combined_df = pd.concat(dfs, ignore_index=True)
    print(combined_df.head())  # Display the first few rows of the combined DataFrame
else:
    print("No tables found.")

```
