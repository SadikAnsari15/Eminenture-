```python
import camelot
import pandas as pd

# Read the PDF and extract tables using 'stream' flavor
tables = camelot.read_pdf('_000011888-0.2.pdf', pages='13-24', flavor='stream')

# Check the number of tables extracted
print(f"Total tables extracted: {len(tables)}")

# Proceed only if tables are found
if len(tables) > 0:
    # Initialize an empty list to store DataFrames
    all_tables = []

    # Iterate through all extracted tables
    for i in range(len(tables)):
        df = tables[i].df

        # Optionally drop empty rows or columns
        df_cleaned = df.dropna(how='all', axis=0)  # Drop rows where all values are NaN
        df_cleaned = df_cleaned.dropna(how='all', axis=1)  # Drop columns where all values are NaN

        # Reset the index
        df_cleaned.reset_index(drop=True, inplace=True)

        # Append the cleaned DataFrame to the list
        all_tables.append(df_cleaned)

    # Concatenate all DataFrames into one
    final_df = pd.concat(all_tables, ignore_index=True)

    # Display the extracted table content
    print("\nExtracted Table Content:\n")
    print(final_df.to_string())  # Print formatted table

    # Save the table to a CSV file (optional)
    final_df.to_csv('extracted_table_all_pages.csv', index=False)
    #print("\nTable saved as 'extracted_table_all_pages.csv'")

    # Show the DataFrame if running in a Jupyter Notebook
    final_df
else:
    print("No tables found.")
```
